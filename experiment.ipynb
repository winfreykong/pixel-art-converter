```python
from skimage import io
from pyxelate import Pyx, Pal
```

## Reading images

```python
img = io.imread('assets/mario.jpg')
img2 = io.imread('assets/luigi.webp')
```

```python
io.imshow(img)
```

## Fit on one image and apply to all other images

```python
downsample=5
palette=10
pyx = Pyx(factor=downsample, palette=palette, dither='none').fit(img)
```

```python
mario = pyx.transform(img)
```

```python
io.imshow(mario)
```

```python
luigi = pyx.transform(img2)
```

```python
io.imshow(luigi)
```

Note how it takes after the color palettes of Mario's
